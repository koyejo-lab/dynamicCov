```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io
```

```python
motor = scipy.io.loadmat('../data/HCP_motor/reg_motor_conv.mat')
motor_conv = motor["reg_motor_conv"]
label= ["Right Hand Tapping", "Left Foot Tapping", "Tongue Wagging", "Right Foot Tapping", "Left Hand Tapping"]
print(motor_conv.shape)

fig, ax = plt.subplots(nrows=5, ncols=1,figsize=(8,6.4))

color_map = ['blue','green','red','black','purple']
for i in range(5):
    activation = np.zeros(motor_conv[:,i].shape)
    activation[np.where(motor_conv[:,i]>0)] = 1
    ax[i].set_ylim(-0.2,1.2)
    ax[i].plot(activation, label=label[i], color=color_map[i])
    ax[i].set_title("Task: "+label[i], fontsize=18,  x=0.5, y=1.2)

#fig.suptitle('Task Activation Map')
#ax[4].set_xlabel('time (s)')
fig.text(0.5,0.04, "time(s)", ha="center", va="center", fontsize=18)
fig.text(0.05,0.5, "0/1 activation", ha="center", va="center", rotation=90, fontsize=18)
plt.subplots_adjust(left=0.1, bottom=0.1, right=None, top=None, wspace=None, hspace=1.5)
plt.savefig('../figures/activation_map.pdf',format='pdf')
#plt.legend(loc='lower right')
```
